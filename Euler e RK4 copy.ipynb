```python
'''
Calcula o valor de 'pi' via Monte Carlo, e plota um gráfico com os pontos dentro e fora do círculo presente no método.

Argumentos
------------------
  i, j: int
      Variaveis de controle de entrada e saida

  aleatorio: Array
      Pontos Dentro e Fora do círculo de raio 1

  Dentro: Lista
      Pontos Dentro do círculo

  Fora: Lista
      Pontos Fora do círculo

  R: float
      Distância de um ponto partindo da origem do gráfico

  pi: float
      Recebe o valor da varíavel pi calculada

  xD, yD: float
      Pontos do gráfico que estão dentro do círculo

  xF, yF: float
      Pontos do gráfico que estão fora do círculo

'''
import numpy as np
import matplotlib.pyplot as plt



def sorteio(N, pontos = False):
    i = 0
    j = 0

    aleatorio = np.random.uniform(0, 1, size=(N,2)) #Gera N pontos "aleatórios"
    Dentro = [] #Definimos a lista vazia de pontos Dentro
    Fora = [] #Definimos a lista vazia de pontos Fora

    for i in range(i, N):
      R = (aleatorio[i, 0]**2 + aleatorio[i, 1]**2)**0.5 #Calcula a distancia da origem ao ponto (Teorema de Pitágoras)
      i += 1
      if R <= 1:
        j += 1
        Dentro.append(aleatorio[i-1]) #Adiciona o ponto na lista Dentro
      else:
        Fora.append(aleatorio[i-1]) #Adiciona o ponto na lista Fora

    pi = 4*(j/N) #Calcula o valor de pi

    if pontos == True:

        xD, yD = zip(*Dentro) #Desacopla a lista 'Dentro' em coordenadas x e y
        xF, yF = zip(*Fora) #Desacopla a lista 'Fora' em coordenadas x e y


        plt.figure()


        # Plotar os pontos dentro do círculo em azul
        plt.scatter(xD, yD, color='blue', label='Dentro do círculo')
        # Plotar os pontos fora do círculo em vermelho
        plt.scatter(xF, yF, color='red', label='Fora do círculo')

        plt.legend()
        plt.title(f'Gráfico com {N} pontos')
        plt.show

        return pi, Dentro, Fora

    else:
        return pi

  

print (sorteio(10, False))
print (sorteio(100, True))
print (sorteio(1000, True))
print (sorteio(10000, True))
```
