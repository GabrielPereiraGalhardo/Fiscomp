```python
#1 - Gaussianas com diferentes larguras e centros
'''
Realiza o plot do grÃ¡fico das funÃ§Ãµes Gaussiana com diferentes larguras e centros

Argumentos
------------------
  x_n : float
      Variavel que representa o valor de x.
  y_n : float
      Exponenciais.
'''

import matplotlib.pyplot as plt #importa a biblioteca com o cÃ³digo que constrÃ³i o grÃ¡fico
import numpy as np #importa a biblioteca com o cÃ³digo que organiza os dados e pontos do grÃ¡fico


x1 = np.arange (-5, 5, 0.1) # Aqui, a variÃ¡vel 'x' recebe valores entre -5 e 5 com 0.1 de espaÃ§o entre eles
y1 = np.exp(-x1**2) # Exponencial Gaussiana

plt.plot(x1, y1) # Realiza o plot de y por x
plt.tight_layout() # organiza o layout do grÃ¡fico
plt.xlabel('x')  # Define o rÃ³tulo do eixo x
plt.ylabel('y')  # Define o rÃ³tulo do eixo y
plt.title('GrÃ¡ficos da funÃ§Ã£o e^(-x^2)')  # Define o tÃ­tulo do grÃ¡fico
plt.grid(True)  # Adiciona uma grade ao grÃ¡fico
plt.show() # Revela e mostra o grÃ¡fico

print("\n")


x2 = np.arange (3, 7, 0.1) # Aqui, a variÃ¡vel 'x' recebe valores entre -5 e 5 com 0.1 de espaÃ§o entre eles
y2 = 50*np.exp((-400*(x2 - 5)**2)) # Exponencial Gaussiana

plt.plot(x2, y2)
plt.tight_layout()
plt.xlabel('x')  # Define o rÃ³tulo do eixo x
plt.ylabel('y')  # Define o rÃ³tulo do eixo y
plt.title('GrÃ¡ficos da funÃ§Ã£o 50*e^(-400*(x - 5)^2)')  # Define o tÃ­tulo do grÃ¡fico
plt.grid(True)  # Adiciona uma grade ao grÃ¡fico
plt.show()

print("\n")



x3 = np.arange (-15, 7, 0.1) # Aqui, a variÃ¡vel 'x' recebe valores entre -5 e 5 com 0.1 de espaÃ§o entre eles
y3 = 2*np.exp((-0.04*(x3 + 5)**2)) # Exponencial Gaussiana

plt.plot(x3, y3)
plt.tight_layout()
plt.xlabel('x')  # Define o rÃ³tulo do eixo x
plt.ylabel('y')  # Define o rÃ³tulo do eixo y
plt.title('GrÃ¡ficos da funÃ§Ã£o 2*e^(-0,04*(x + 5)^2)')  # Define o tÃ­tulo do grÃ¡fico
plt.grid(True)  # Adiciona uma grade ao grÃ¡fico
plt.show()
```

```python
#1 - funÃ§Ã£o exponencial
'''
Realiza o plot do grÃ¡fico da FunÃ§Ã£o exponencial com diferentes valores de a
'''
import matplotlib.pyplot as plt
import numpy as np
a = [-10, -1, 2, 10]

n = 0

for n in range(len(a)):
    x = np.arange (0, 5, 0.1)
    y = np.exp(a[n]*x) # FunÃ§Ã£o exponencial

    plt.plot(x, y)
    plt.tight_layout()
    plt.xlabel('x')  # Define o rÃ³tulo do eixo x
    plt.ylabel('y')  # Define o rÃ³tulo do eixo y
    plt.title(f'GrÃ¡ficos da FunÃ§Ã£o exp({a[n]}x)')  # Define o tÃ­tulo do grÃ¡fico
    plt.grid(True)  # Adiciona uma grade ao grÃ¡fico
    plt.show()
```

```python
#2 - Bhaskara
import matplotlib.pyplot as plt
import numpy as np

def bhaskara(a, b, c):
    '''
    Calcula e retorna as raÃ­zes de um polinÃ´mio de segundo grau utilizando a fÃ³rmula de Bhaskara

    Argumentos
    -----------
    delta: float
        Valor do discriminante.

    a, b, c: float
        Coeficientes do polinÃ´mio.

    Retorna
    -----------
    x1, x2: float
        As raÃ­zes do polinÃ´mio.

    '''

    if a == 0 :
      x1 = -c/b
      x2 = -c/b
      return x1, x2


    delta = b**2 - 4*a*c  # valor do discriminante

    x1 = (-b + delta**0.5) / (2*a) # primeira raiz
    x2 = (-b - delta**0.5) / (2*a) # segunda raiz

    return x1, x2 # tupla com as duas raÃ­zes

def f(x, a, b, c):
    '''
    Calcula e retorna o valor de uma funÃ§Ã£o polinomial de segundo grau
     Argumentos
    -----------
    x: float
        Ponto em que a funÃ§Ã£o Ã© avaliada.
    a, b, c: float
        Coeficientes do polinÃ´mio.
    '''
    return a*x**2 + b*x + c

x1, x2 = bhaskara(1, 2, 3)
print(f'A primeira raiz Ã©: {x1:.4f}')
print(f'A segunda raiz Ã©: {x2: .4f}')
print("Caso 1: Discriminante negativo")
x = np.arange (-8, 6, 0.1)
y = x**2 + 2*x + 3

plt.plot(x, y, label ='Parte real')
plt.axhline(y=0, color='black', linewidth=1) #Evidencia o eixo y = 0
plt.axvline(x=0, color='black', linewidth=1) #Evidencia o eixo x = 0
plt.tight_layout()
plt.xlabel('x')  # Define o rÃ³tulo do eixo x
plt.ylabel('y')  # Define o rÃ³tulo do eixo y
plt.title(f'GrÃ¡fico da parte real da funÃ§Ã£o x**2 + 2*x + 3')  # Define o tÃ­tulo do grÃ¡fico
plt.grid(True)  # Adiciona uma grade ao grÃ¡fico
plt.show()

print("\n")

x1, x2 = bhaskara(1, 4, 4)
print('A primeira raiz Ã©:', x1)
print('A segunda raiz Ã©:', x2)
print("Caso 2: Discriminante igual Ã  0")
x = np.arange (-10, 6, 0.1)
y = x**2 + 4*x + 4

plt.plot(x, y)
plt.scatter([x1, x2], [f(x1, 1, 4, 4), f(x2, 1, 4, 4)], color='red', marker='o')
plt.axhline(y=0, color='black', linewidth=1)
plt.axvline(x=0, color='black', linewidth=1)
plt.tight_layout()
plt.xlabel('x')  # Define o rÃ³tulo do eixo x
plt.ylabel('y')  # Define o rÃ³tulo do eixo y
plt.title(f'GrÃ¡fico da parte real da funÃ§Ã£o 2*x + 3')  # Define o tÃ­tulo do grÃ¡fico
plt.grid(True)  # Adiciona uma grade ao grÃ¡fico
plt.show()

print("\n")

x1, x2 = bhaskara(2, 12, -14)
print('A primeira raiz Ã©:', x1)
print('A segunda raiz Ã©:', x2)
print("Caso 1: Discriminante positivo")
x = np.arange (-12, 6, 0.1)
y = 2*x**2 + 12*x + -14

plt.plot(x, y)
plt.scatter([x1, x2], [f(x1, 2, 12, -14), f(x2, 2, 12, -14)], color='red', marker='o')
plt.axhline(y=0, color='black', linewidth=1)
plt.axvline(x=0, color='black', linewidth=1)
plt.tight_layout()
plt.xlabel('x')  # Define o rÃ³tulo do eixo x
plt.ylabel('y')  # Define o rÃ³tulo do eixo y
plt.title(f'GrÃ¡fico da parte real da funÃ§Ã£o 2*x + 3')  # Define o tÃ­tulo do grÃ¡fico
plt.grid(True)  # Adiciona uma grade ao grÃ¡fico
plt.show()
```
