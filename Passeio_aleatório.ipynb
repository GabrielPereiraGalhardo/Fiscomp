```python
import numpy as np
import matplotlib.pyplot as plt


"""
Plota um gráfico da posição do caminhante, no problema do passeio aleatório, em função do número de passos N

Args:
  position: Lista
    Armezena valores que indicam se o bebado andou para esquerda ou para a direita

  i: Int
    Parâmetro que controla o loop 'for'

  step: Int
    Informa para onde o caminhante irá andar, para esquerda ou para a direita.

  c: Float
    Armazena um número aleatório gerado pela função rng.random()

  p, p1, p2 : Float
    Probabilidade do caminhante andar para esquerda ou para direita

  N: Int
    Número de passos do caminhante
      
  passos1, passos2: Int
    Armazena a quantidade de passos dados pelo caminhante

  teste1, teste2: Lista
    Armazena as posições que o caminhante se encontrou em cada passo dado
        

""" 



rng = np.random.default_rng()

def random_walk(N, p):

    position = [0]
    i = 0
    step = 0

    for i in range(N):
        c = rng.random()
        if c < p:
          step += 1
        else:
          step -= 1

        position.append(step)
        i += 1

    return position


N = 1000
p1 = 0.5
teste1 = random_walk(N, p1)
passos1 = np.linspace(0, N, N + 1)
plt.plot(passos1, teste1, 'b', markersize=0.3, label = 'p = 0.5')

N = 1000
p2 = 0.6
teste2 = random_walk(N, p2)
passos2 = np.linspace(0, N, N + 1)

plt.plot(passos2, teste2, 'r', markersize=0.3,  label = 'p = 0.8')

plt.xlabel('Passos')
plt.ylabel('Posição')
plt.title(f'Caminhar do bebado com probabilidade p')
plt.legend()
plt.show()



```

```python
'''
Roda o código acima 1000 vezes para diferentes valores de probabilidade, e plota um gráfico da posição em função dos passos

Args:
  posicoes_iniciais_p1, posicoes_iniciais_p2: Lista
    Armazena a posição do primeiro passo dado pelo caminhante em probabilidades específicas

  posicoes_meio_p1, posicoes_meio_p2: Lista
    Armazena a posição do passo que define a metade de todos os passos dados pelo caminhante em probabilidades específicas

  posicoes_finais_p1, posicoes_finais_p2: Lista
    Armazena a posição do ultimo passo dado pelo caminhante em probabilidades específicas

  j: Int
    Parâmetro de controle do loop 'for'

'''



posicoes_iniciais_p1 = []
posicoes_meio_p1 = []
posicoes_finais_p1 = []

posicoes_iniciais_p2 = []
posicoes_meio_p2 = []
posicoes_finais_p2 = []



for j in range(1000):

  N = 1000
  p1 = 0.5
  teste1 = random_walk(N, p1)
  passos1 = np.linspace(0, N, N + 1)

  posicoes_iniciais_p1.append(teste1[0])  # Posição no início (sempre 0)
  posicoes_meio_p1.append(teste1[N // 2])  # Posição no meio
  posicoes_finais_p1.append(teste1[-1])   # Posição no final

  plt.plot(passos1, teste1, 'b', alpha =0.1)


  N = 1000
  p2 = 0.6
  teste2 = random_walk(N, p2)
  passos2 = np.linspace(0, N, N + 1)

  posicoes_iniciais_p2.append(teste2[0])  # Posição no início (sempre 0)
  posicoes_meio_p2.append(teste2[N // 2])  # Posição no meio
  posicoes_finais_p2.append(teste2[-1])   # Posição no final

  plt.plot(passos2, teste2, 'r', alpha = 0.1)

  j += 1

plt.xlabel('Passos')
plt.ylabel('Posição')
plt.title(f'Caminhar do bebado com probabilidade p')
plt.legend()
plt.show()



```

```python
'''
Calcula a média e o desvio padrão da posição do caminhante em função do número de passos.
Plota o gráfico da média e do desvio padrao em diferentes probabilidades.

Arg:
  posições1, posições2: Lista
    Armazena as posições de cada passo dado pelo caminhante em probabilidades específicas
  
  k: Int
    Parâmetro de controle do loop 'for' 

  media1, media2: Float
    Armazena a média dos passos dados em cada passeio realizado no código acima

  desvio1, desvio2: Float
    Armazena o desvio padrão dos passos dados em cada passeio realizado no código acima


'''


posições1 = []
posições2 = []

for k in range(100):
  posições1.append(random_walk(N,p1))
  posições2.append(random_walk(N,p2))

media1 = np.mean(posições1, axis = 0)
desvio1 = np.std(posições1, axis = 0)
media2 = np.mean(posições2, axis = 0)
desvio2 = np.std(posições2, axis = 0)

plt.plot(passos1, media1, 'b', label = 'p = 0.5')
plt.plot(passos2, media2, 'r', label = 'p = 0.8')
plt.xlabel('Passos')
plt.ylabel('Posição')
plt.title('Média da posição em função dos passos')
plt.legend()
plt.show()

plt.plot(passos1, desvio1, 'b', label = 'p = 0.5')
plt.plot(passos2, desvio2, 'r', label = 'p = 0.6')
plt.xlabel('Passos')
plt.ylabel('Desvio')
plt.title('Desvio padrão da posição em função dos passos')
plt.legend()
plt.show()
```
