```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng()

def random_walk(N, p):

    position = [0]
    i = 0
    step = 0

    for i in range(N):
        c = rng.random()
        if c < p:
          step += 1
        else:
          step -= 1

        position.append(step)
        i += 1

    return position


N = 1000
p1 = 0.5
teste1 = random_walk(N, p1)
passos1 = np.linspace(0, N, N + 1)
plt.plot(passos1, teste1, 'b', markersize=0.3, label = 'p = 0.5')

N = 1000
p2 = 0.6
teste2 = random_walk(N, p2)
passos2 = np.linspace(0, N, N + 1)

plt.plot(passos2, teste2, 'r', markersize=0.3,  label = 'p = 0.8')

plt.xlabel('Passos')
plt.ylabel('Posição')
plt.title(f'Caminhar do bebado com probabilidade p')
plt.legend()
plt.show()



```

```python
for j in range(100):

  N = 1000
  p1 = 0.5
  teste1 = random_walk(N, p1)
  passos1 = np.linspace(0, N, N + 1)

  plt.plot(passos1, teste1, 'b', alpha =0.1)

  N = 1000
  p2 = 0.6
  teste2 = random_walk(N, p2)
  passos2 = np.linspace(0, N, N + 1)

  plt.plot(passos2, teste2, 'r', alpha = 0.1)

  j += 1

plt.xlabel('Passos')
plt.ylabel('Posição')
plt.title(f'Caminhar do bebado com probabilidade p')
plt.legend()
plt.show()



```
