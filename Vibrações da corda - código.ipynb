```
import numpy as np
import matplotlib.pyplot as plt

'''
 Resolve numericamente a equação de ondas unidimensional com condições de Dirichlet
    utilizando o método de diferenças finitas.

    Argumentos:
    L : int
        Comprimento da barra.

    N : int
        Número de pontos internos da matriz.

    dx : float
        Espaçamento entre os pontos da matriz.

    diagonal : ndarray
        Vetor principal da matriz tridiagonal.

    n_diagonal : ndarray
        Vetores superiores e inferiores da matriz tridiagonal.

    A : ndarray
        Matriz tridiagonal (N x N) que representa o operador d²/dx².

    lambdas : ndarray
        Autovalores de A, correspondem aos valores de λ da equação.

    phis : ndarray
        Autovetores normalizados de A, representam os modos normais da equação de ondas.

    x : ndarray
        Coordenadas espaciais correspondentes aos N pontos internos da barra.
    '''

L = 1.0
N = 100
dx = L / (N + 1)


diagonal = -2 * np.ones(N)
n_diagonal = np.ones(N - 1)
A = (np.diag(diagonal) + np.diag(off_diagonal, k=1) + np.diag(off_diagonal, k=-1)) / dx**2


lambdas, phis = np.linalg.eigh(A)


x = np.linspace(dx, L - dx, N)

plt.figure(figsize=(10, 6))
for i in range(4):
    plt.plot(x, phis[:, i], label=f'Modo {i+1}, λ = {lambdas[i]:.2f}')
    plt.title(f'Modo normal {i+1}')
    plt.xlabel('x')
    plt.ylabel(r'$\phi(x)$')
    plt.grid(True)
    plt.legend()
    plt.show()
```
