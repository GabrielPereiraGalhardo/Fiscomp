```python
"""
Calcula e plota o mapa logstico para uma faixa de valores de r.

Args:
    r_valores: Lista de valores de r para plotar.
    x0: Valor inicial de x.
    N: Nmero de iteraes.
"""

import numpy as np
import matplotlib.pyplot as plt

def mapa_logistico(r_valores, x0, N): # Changed 'r' to 'r_valores' to match function call and docstring

    plt.figure()

    for r in r_valores:
        x_valores = [x0]
        x_n = x0
        n_valores = [0]

        for n in range(1, N):
            x_prox = r * x_n * (1 - x_n)
            x_valores.append(x_prox)
            x_n = x_prox
            n_valores.append(n)

        plt.plot(n_valores, x_valores, 'o-', ms=1)
        plt.xlabel('n')
        plt.ylabel('x')
        plt.title(f'Mapa Logstico para r = {r}')
        plt.show()


    return x_valores

r_valores = [2, 3.3, 3.5, 3.9]

print(mapa_logistico(r_valores, 0.5, 100)) # Changed 'r' to 'r_valores' to match function definition
```

```python
"""
Calcula e plota o mapa logstico para uma faixa de valores de r.

Args:
    r_valores: Lista de valores de r para plotar.
    x0: Valor inicial de x.
    N: Nmero de iteraes.
"""

import numpy as np
import matplotlib.pyplot as plt

def mapa_logistico(r_valores, x0, N): # Changed 'r' to 'r_valores' to match function call and docstring

    plt.figure()

    M = 1000

    for r in r_valores:
        x_valores = [x0]
        x_n = x0

        for n in range(1, N):
            x_prox = r * x_n * (1 - x_n)
            x_valores.append(x_prox)
            x_n = x_prox
        
        M = int(N/2)
        plt.plot(r*np.ones_like(x_valores[-M:]), x_valores[-M:], 'k.', markersize=0.5)

    plt.xlabel('r')
    plt.ylabel('x')
    plt.title(f'Mapa Logstico com {N} pontos')
    plt.show()


    return x_valores

r_valores = np.arange(1, 4, 0.1)

print(mapa_logistico(r_valores, 0.5, 100))
print(mapa_logistico(r_valores, 0.5, 500))
print(mapa_logistico(r_valores, 0.5, 1000))
print(mapa_logistico(r_valores, 0.5, 2000)) 
```
