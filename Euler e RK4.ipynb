```python
'''
    Calcula os valores de tempo e suas respectivas velocidades para a equação diferencial de um corpo em queda livre, com resistência do ar, 
    utilizando o método de Euler.

    Argumentos
    ----------
    n: int
        Parâmetro que controla o loop 'for'.

    g: float
        Gravidade.

    b = float
        Coeficiente de arrasto (resistência do ar).
    
    m = int
        Massa do corpo em queda livre.

    t = array
        Armazena diferentes valores de tempo.

    v = array
        Armazena diferentes valores de velocidade.

    tn = float
        Próximo valor de tempo da formula iterativa.

    vn = float
        Próximo valor de velocidade da formula iterativa.

    h = float
        Variação de tempo entre um valor de t e o próximo.

    Retorna
    -------
    t e v = Lista
        Valores do tempo e valores das velocidades nesses tempos.
    
    Também retorna um gráfico de v(t), onde é possível analisar a velocidade terminal.
    '''


import numpy as np 
import matplotlib.pyplot as plt

n = 0
g = 9.8
b = 0.1
m = 1

def Euler(h):
    t = np.array([0])
    v = np.array([0])

    for n in range(100):
        tn = t[-1] + 0.5
        t = np.append(t, tn)
        vn = v[-1] + h*(-g -(b/m)*v[-1])
        v = np.append(v, vn)

    return t, v


plt.plot(Euler(0.5)[0], Euler(0.5)[1], 'g', markersize=0.3, label = 'Δt = 0.5')
plt.xlabel('Tempo')
plt.ylabel('Velocidade')
plt.title('Método de Euler')
plt.legend()
plt.show()
```

```python

'''
    Calcula os valores de tempo e suas respectivas velocidades para a equação diferencial de um corpo em queda livre, com resistência do ar, 
    utilizando o método de Runge-Kutta de quarta ordem.

    Argumentos
    ----------
    n: int
        Parâmetro que controla o loop 'for'.

    g: float
        Gravidade.

    b = float
        Coeficiente de arrasto (resistência do ar).
    
    m = int
        Massa do corpo em queda livre.

    t = array
        Armazena diferentes valores de tempo.

    v = array
        Armazena diferentes valores de velocidade.

    vn = float
        Próximo valor de velocidade da fórmula de Runge-Kutta.

    k1, k2, k3 e k4 = float
        Parâmetros provindos do método de Runge-Kutta de quarda ordem. São utilizados para calcular
        os valores de vn.
  
    h = float
        Variação de tempo entre um valor de t e o próximo.

    Retorna
    -------
    t e v = Lista
        Valores do tempo e valores das velocidades nesses tempos.
    
    Também retorna um gráfico de v(t), onde é possível analisar a velocidade terminal.
    '''




def RK4(h):
    n = 0
    g = 9.8
    b = 0.1
    m = 1
    t = np.array([0])
    v = np.array([0])

    for n in range(100):
        k1 = -g -(b*v[-1])/m
        k2 = -g -(b*(v[-1] + (h*k1)/2))/m 
        k3 = -g -(b*(v[-1] + (h*k2)/2))/m
        k4 = -g -(b*(v[-1] + (h*k3)))/m

        vn = v[-1] + (h*(k1 + 2*k2 + 2*k3 + k4))
        v = np.append(v, vn)
        t = np.append(t, t[-1] + h)
    
    
    return t, v
    
plt.plot(RK4(0.1)[0], RK4(0.1)[1], 'r', markersize=0.3, label = 'Δt = 0.1')
plt.plot(RK4(0.5)[0], RK4(0.5)[1], 'b', markersize=0.3, label = 'Δt = 0.5')
plt.plot(RK4(1.0)[0], RK4(1.0)[1], 'g', markersize=0.3, label = 'Δt = 1.0')
plt.title('RK4')
plt.xlabel('Tempo')
plt.xlim(-1, 10)
plt.ylabel('Velocidade')
plt.legend()
plt.show()



```

```python
'''
Comparação gráfica entre o método de Euler e o de Runge-Kutta de quarta ordem.
'''


plt.plot(RK4(0.5)[0], RK4(0.5)[1], 'b', markersize=0.3, label = 'RK4')
plt.plot(Euler(0.5)[0], Euler(0.5)[1], 'r', markersize=0.3, label = 'Euler')
plt.title('Δt = 0.5')
plt.xlabel('Tempo')
plt.ylabel('Velocidade')
plt.legend()
plt.show()
```
